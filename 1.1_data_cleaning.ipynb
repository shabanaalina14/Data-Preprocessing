### types of method for removing outliers

1.z-score method: calculate std score of each data points ,z-score greater thn 3 of less thn -3 considered outliers

2.IQR method : IQR above 3rd quantile or below 1st considered as outliers

3.percentile method : remove data points that falls outside a certain percentile.

# 2.IQR method

```python
sns.boxplot(data['Diabetes_Pedigree'])
```

```python
q1=data['Diabetes_Pedigree'].quantile(0.25)
q3=data['Diabetes_Pedigree'].quantile(0.75)
iqr=q3-q1
df=data[(data['Diabetes_Pedigree']>=q1-1.5*iqr) & (data['Diabetes_Pedigree']<=q3+1.5*iqr)]
sns.boxplot(df['Diabetes_Pedigree'])
```
